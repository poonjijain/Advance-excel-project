```python
def pos(text,pw):
  list1_lower = [str(x).lower() for x in text]
  list2_lower = [str(x).lower() for x in pw]
  matching = [x for x in list1_lower if x in list2_lower]
  return (len(matching))
```

**Negative words count**

```python
negwords=pd.read_excel("/content/drive/MyDrive/BlackCoffee.xlsx",sheet_name="Sheet3",usecols=["Negwords"])
nw=negwords["Negwords"].dropna().tolist()
print(nw)
```

```python
def neg(text,nw):
  list1_lower = [str(x).lower() for x in text]
  list2_lower = [str(x).lower() for x in nw]
  matching = [x for x in list1_lower if x in list2_lower]
  return (len(matching))
```

**Polarity count and subjectivity count**

```python
def polarity_subjectivity(pos,neg,sw):
  ps=pos-neg/((pos+neg)+0.000001)
  sub=pos-neg/((sw)+0.000001)
  return(ps,sub)
```

**Analysis of readability**

Average Sentence Length or Average Number of Words Per Sentence


```python
def count(text):
  sentence_token = []
  avg_sent_len=0
  word_token = []
  if len(text) > 0:
    word_token=nltk.word_tokenize(text)
    sentence_token=nltk.sent_tokenize(text)
    if len(sentence_token)>0:
      avg_sent_len=len(word_token)/len(sentence_token)
  return(avg_sent_len)
```

**Words Count**

# Removing any punctuations like ? ! , .

```python
def substitute(text):
  after=[]
  for w in text:
    res=re.sub(r'[^\w\s]',"",w) # not equal to alphanumeric symbols
    if res !="":
      after.append(res)
    
  return after
  
```

Removing the stop words

```python
stopwords=pd.read_excel("/content/drive/MyDrive/BlackCoffee.xlsx",sheet_name="Sheet2",usecols=["Stopwords"])
sw=stopwords["Stopwords"].dropna().tolist()
print(sw)
```

```python
def eliminate(stopwords,clean_data):
  eliminated=[]
  for i in clean_data:
    c=0
    for x in stopwords:
      if str(i).lower() == str(x).lower():
        c=c+1
        continue
    if c==0:
        eliminated.append(i)
  return eliminated
```

**Syllable per word**

```python
def stemming(k):
  lemmatizer = WordNetLemmatizer()
  lem_list = []
  for word in k:
    if word.endswith("ed"):
        lemma = lemmatizer.lemmatize(word, pos='v')  # Assuming it's a verb
    elif word.endswith("es"):
        lemma = lemmatizer.lemmatize(word, pos='n')  # Assuming it's a noun
    else:
        lemma = word 
    
    lem_list.append(lemma)
  return(lem_list)
```

**Complex words count**

```python
def syllable(words):
  c=0
  syl=0
  vowel=["a","i","e","o","u"]
  pattern = r"\b(" + "|".join(vowel) + r")\b"
  for i in words:
    match=0
    for j in i:
      if j in vowel:
        syl+=1
        match+=1
    if match>2:
      c+=1
  return (c,syl)
  
```

**Counting Personal pronouns**


```python

def pp(text):
  words = ["I", "we", "my", "ours", "us", "i", "We", "My", "Ours", "Us", "WE", "OURS", "MY"]
  word_counts=0
  pattern = r"\b(" + "|".join(words) + r")\b"
  matches = re.findall(pattern, text)
  for word in words:
    word_counts =word_counts + matches.count(word) 
  return(word_counts)
```
